# Try differente ML approaches

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

## Approach 1: both ARGs and SNPs

```python
y = pd.merge(args_df["sample_name"], amr_labels_df, left_on="sample_name", right_on="SampleID")
y = y.drop("SampleID", axis=1).set_index("sample_name")
y = y.dropna()
y.apply(pd.value_counts)
```

```python
X = pd.merge(args_df, snps_df, how="left", left_on="sample_name", right_on="SAMPLE_ID")
X = pd.merge(y.reset_index()[["sample_name"]], X, how="left", on="sample_name")
X = X.fillna(0).drop("SAMPLE_ID", axis=1).set_index("sample_name")
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, y_pred, target_names=y.columns))
```

```python
# Plot top 20 features importances
forest_importances = pd.Series(rf.feature_importances_, index=X.columns)
forest_importances = forest_importances.sort_values(ascending=False)[:20]

fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
ax.set_title("Top 20 feature importances")
fig.tight_layout()
```

```python
# Plot bottom 20 features importances
forest_importances = pd.Series(rf.feature_importances_, index=X.columns)
forest_importances = forest_importances.sort_values(ascending=False)[-20:]

fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
ax.set_title("Bottom 20 feature importances")
fig.tight_layout()
```

## Random Forest


We have already tried this model with default parameters but Random Forest has multiple hyperparameters to tune. First of all, we are going to choose some values for each parameter.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation,
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

### Evaluation metrics for each antibiotic (ROC plot + table)

```python
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_predict
```

```python
# Instantiate the random forest classifier
rf = RandomForestClassifier(
    n_estimators=600,
    min_samples_split=10,
    min_samples_leaf=1,
    max_features='sqrt',
    max_depth=110,
    bootstrap=False
)
```

```python
# Create an empty dictionary to store the results
results = {}

# Define the scoring metrics
scoring = ("accuracy", "f1_micro", "roc_auc")

# Perform cross-validation for each class
for class_index, class_name in enumerate(y_train.columns):
    y_true_class = y_train.iloc[:, class_index]

    # Perform cross-validation and calculate scores
    scores = cross_validate(rf, X_train, y_true_class, cv=5, scoring=scoring)

    # Extract the mean and standard deviation of the scores
    mean_scores = {metric + '_mean': np.mean(scores[f'test_{metric}']) for metric in scoring}
    std_scores = {metric + '_std': np.std(scores[f'test_{metric}']) for metric in scoring}

    # Combine the mean and standard deviation scores
    results[class_name] = {**mean_scores, **std_scores}

# Create a DataFrame to display the results
cv_results_by_antibiotic = pd.DataFrame(results).T
cv_results_by_antibiotic
```

```python
# Plot ROC curves for each class
# In order to calculate the True positive rate (TPR) and the False Positive Rate (FPR), we need to get the probabilities scores.
fig, ax = plt.subplots(figsize=(12, 9))
for class_index, class_name in enumerate(y_train.columns):
    rf.fit(X_train, y_train.iloc[:, class_index])
    y_score_class = rf.predict_proba(X_test)[:, 1]
    y_true_class = y_test.iloc[:, class_index]

    # Compute ROC curve and get mean data
    fpr, tpr, _ = roc_curve(y_true_class, y_score_class)
    roc_auc = cv_results_by_antibiotic.loc[class_name, "roc_auc_mean"]
    std = cv_results_by_antibiotic.loc[class_name, "roc_auc_std"]

    # Plot ROC curve
    plt.plot(fpr, tpr, label=f"{class_name} (AUC = {roc_auc:.2f} ± {std:.2f})")

# Set plot properties
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right", fontsize="small")
plt.show()
```
